# MADALINE

```python
import numpy as np

# Function to initialize weights for each Adaline unit
def initialize_weights(input_dim):
    return np.zeros(input_dim + 1)  # Including the bias term

# Linear activation function for each Adaline unit
def activation_function(x):
    return x

# Predict output for an Adaline unit
def predict(weights, X):
    return activation_function(np.dot(X, weights[1:]) + weights[0])

# Update weights for each Adaline unit
def update_weights(weights, X, y, learning_rate):
    output = predict(weights, X)
    errors = y - output
    weights[1:] += learning_rate * np.dot(X.T, errors)  # Update feature weights
    weights[0] += learning_rate * errors.sum()          # Update bias
    return weights

# Train a single Adaline unit
def adaline_train(X, y, learning_rate=0.01, epochs=50):
    weights = initialize_weights(X.shape[1])
    for _ in range(epochs):
        weights = update_weights(weights, X, y, learning_rate)
    return weights

# Train multiple Adaline units (Madaline)
def madaline_train(X, y, num_units=3, learning_rate=0.01, epochs=50):
    # Train multiple Adaline units
    adaline_weights = []
    for unit in range(num_units):
        print(f"Training Adaline unit {unit + 1}")
        unit_weights = adaline_train(X, y, learning_rate, epochs)
        adaline_weights.append(unit_weights)
    return adaline_weights

# Combine outputs of all Adaline units (simple majority voting mechanism)
def madaline_predict(weights_list, X):
    predictions = []
    for weights in weights_list:
        predictions.append(predict(weights, X))
    predictions = np.array(predictions)
    # Final prediction: Average of all unit outputs
    return np.mean(predictions, axis=0) >= 0.5  # Threshold at 0.5

# Example dataset
X = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])  # Input features
y = np.array([1, 1, 0, 0])                      # Target labels

# Train the Madaline model (with 3 Adaline units)
madaline_weights = madaline_train(X, y, num_units=3, learning_rate=0.01, epochs=100)

# Predict using the trained Madaline model
final_predictions = madaline_predict(madaline_weights, X)
print("Madaline Predictions:", final_predictions)
```
