```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Provided data points and their corresponding classes
data_points = [(1, 1), (1, 3), (2, 4), (3, 5), (4, 4), (3, 2), (4, 1), (5, 2)]
classes = ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B']

# Features and labels
X = [[x[0], x[1]] for x in data_points]
y = classes

# Initialize and fit LDA model
lda = LinearDiscriminantAnalysis()
lda.fit(X, y)

# New data point
new_data_point = [[1, 4]]

# Predict class for the new data point
predicted_class = lda.predict(new_data_point)

print("Predicted class for (1, 4):", predicted_class[0])
```
