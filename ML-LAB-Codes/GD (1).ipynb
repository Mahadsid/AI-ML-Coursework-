```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
```

```python
def y_func(x):
  return x ** 3 -  10 *x**2

def y_derv(x):
  return 3 * x **2 - 200 * x
```

```python
X = np.arange(-100, 100, 0.1)
Y = y_func(X)
```

```python
plt.plot(X, Y)
plt.show()
```

```python
current_pos = (100, y_func(100))
```

```python
plt.plot(X, Y)
plt.scatter(current_pos[0], current_pos[1], color="red")
plt.show()
```

```python
learning_rate = 0.01
for _ in range(500):
  new_x = current_pos[0] - learning_rate * y_derv(current_pos[0])
  new_y = y_func(new_x)
  current_pos = (new_x, new_y)
  print(new_x, new_y)

  plt.plot(X, Y)
  plt.scatter(current_pos[0], current_pos[1], color="red")
  plt.pause(0.0001)
  plt.clf()
```
